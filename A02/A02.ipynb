Como forma de contornar esse problema, diversas estratégias podem ser utilizadas para normalização dessa característica. Cite e explique uma dessas estratégias.


---


**R**: Uma estratégia que pode ser utilizada para contornar o problema da variação de luminosidade é a normalização da imagem (input). Para isso, faz-se a subtração pela média e depois a normalização pelo desvio padrão da imagem, para cada uma das imagens. Assim, temos que:  $$f(I(x)) =  [ I(x) - μ ] / σ$$ , onde μ se refere à média da imagem e σ ao desvio padrão. Ao final, fazemos a subtração de cada pixel das imagens finais e elevamos ao quadrado $$\sum_{x\in \Omega}(f(I(x))-f(J(x)))² = valor baixo,$$ onde I é a primeira imagem e J é a segunda imagem.

```python
import cv2
from google.colab.patches import cv2_imshow

# Carregar a imagem médica
image = cv2.imread('anomalia0.jpg', cv2.IMREAD_GRAYSCALE)

# Mostrar a imagem com os keypoints
cv2_imshow(image)
```

```python
# Inicializar o detector
orb = cv2.ORB_create()

# Configurar o threshold
orb.setMaxFeatures(10)

# Detecta os keypoints
keypoints, descriptors = orb.detectAndCompute(image, None)

# Desenha os keypoints na imagem
image_with_keypoints = cv2.drawKeypoints(image, keypoints, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# Mostrar a imagem com os keypoints
cv2_imshow(image_with_keypoints)
```
