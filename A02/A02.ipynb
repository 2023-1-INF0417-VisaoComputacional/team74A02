##Questão Prática:

Você está desenvolvendo um sistema de detecção de keypoints em imagens médicas para auxiliar no diagnóstico de doenças. Considerando a necessidade de robustez e invariância a rotações e escalas, qual dos seguintes algoritmos de detecção e descrição de keypoints você utilizaria: SURF, SIFT ou ORB? Explique sua escolha e forneça um exemplo de código em Python que ilustre como usar o algoritmo escolhido para extrair os keypoints e as correspondentes descritores de uma imagem médica.


---



```python
import cv2
from google.colab.patches import cv2_imshow
# Carregar a imagem médica
image = cv2.imread('bala2.jpg', cv2.IMREAD_GRAYSCALE)

# Inicializar o detector SIFT
sift = cv2.SIFT_create(contrastThreshold=0.09)

# Detectar os keypoints na imagem
keypoints = sift.detect(image, None)

# Calcular os descritores dos keypoints
keypoints, descriptors = sift.compute(image, keypoints)

# Desenhar os keypoints na imagem
image_with_keypoints = cv2.drawKeypoints(image, keypoints, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# Mostrar a imagem com os keypoints
cv2_imshow(image_with_keypoints)
```
